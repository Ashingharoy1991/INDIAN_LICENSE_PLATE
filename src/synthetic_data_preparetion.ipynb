###  Script: Check for Same File Names in Two Directories

```python
import os

# Set your two directory paths
dir1 = '/home/arijit/LPR_project/synthetic_data_generated/data_v001/final_data'
dir2 = '/home/arijit/LPR_project/synthetic_data_generated/data_v001/en_train_filtered'

def find_common_files(directory1, directory2):
    files1 = set(os.listdir(directory1))
    files2 = set(os.listdir(directory2))

    # Find common file names
    common_files = files1.intersection(files2)

    if common_files:
        print(f"Common files found ({len(common_files)}):")
        for file in sorted(common_files):
            print(file)
    else:
        print("No common files found.")

# Run the function
find_common_files(dir1, dir2)
```

### Finds common file names between dir1 and dir2

Skips the common files

Copies the remaining (non-common) files from dir1 into a new folder (e.g., only_in_dir1)

```python
import os
import shutil

# Set your two directory paths

dir1 = '/home/arijit/LPR_project/synthetic_data_generated/data_v001/en_train_filtered'
dir2 = '/home/arijit/LPR_project/synthetic_data_generated/data_v001/final_data'
output_dir = '/home/arijit/LPR_project/synthetic_data_generated/data_v001/only_in_dir1'

def move_non_common_files(src_dir, compare_dir, target_dir):
    os.makedirs(target_dir, exist_ok=True)

    files_src = set(os.listdir(src_dir))
    files_cmp = set(os.listdir(compare_dir))

    # Find common file names
    common_files = files_src.intersection(files_cmp)
    remaining_files = files_src - common_files

    print(f"Total files in {src_dir}: {len(files_src)}")
    print(f"Common files: {len(common_files)}")
    print(f"Files only in {src_dir}: {len(remaining_files)}")

    for file in remaining_files:
        src_path = os.path.join(src_dir, file)
        dst_path = os.path.join(target_dir, file)
        shutil.copy(src_path, dst_path)

    print(f"Copied {len(remaining_files)} non-common files to '{target_dir}'.")

# Run the function
move_non_common_files(dir1, dir2, output_dir)
```

```python
import os
import shutil
import random

def is_image_file(filename):
    image_extensions = {'.jpg', '.jpeg', '.png', '.bmp', '.gif', '.tiff'}
    return os.path.splitext(filename)[1].lower() in image_extensions

def copy_random_and_fixed_images(src1, src2, target_dir):
    os.makedirs(target_dir, exist_ok=True)

    # List image files in both directories
    src1_images = [f for f in os.listdir(src1) if is_image_file(f)]
    src2_images = [f for f in os.listdir(src2) if is_image_file(f)]

    num_images_to_sample = len(src2_images)
    print(f"Images in src2: {num_images_to_sample}")

    if num_images_to_sample > len(src1_images):
        raise ValueError("src1 does not have enough images to sample.")

    # Randomly select images from src1
    sampled_src1_images = random.sample(src1_images, num_images_to_sample)

    # Copy src2 images
    for img in src2_images:
        shutil.copy(os.path.join(src2, img), os.path.join(target_dir, f"src2_{img}"))

    # Copy sampled src1 images
    for img in sampled_src1_images:
        shutil.copy(os.path.join(src1, img), os.path.join(target_dir, f"src1_{img}"))

    print(f"Copied {num_images_to_sample} images from src1 and src2 to '{target_dir}'.")

# Example usage
src1 = "/home/arijit/LPR_project/synthetic_data_generated/data_v001/en_train_filtered"
src2 = "/home/arijit/LPR_project/synthetic_data_generated/data_v001/skewed_dataset"
target_dir = "/home/arijit/LPR_project/synthetic_data_generated/data_v001/final_data"

copy_random_and_fixed_images(src1, src2, target_dir)
```

### Script: Count and Move Images

```python
import os
import shutil

# Set your source and destination paths
source_dir = '/home/arijit/LPR_project/synthetic_data_generated/data_v001/final_data'
destination_dir = '/home/arijit/LPR_project/synthetic_data_generated/data_v001/skewed_dataset'

# Supported image extensions
image_extensions = {'.jpg', '.jpeg', '.png', '.bmp', '.gif', '.tiff'}

# Count and move image files
def count_and_move_images(src, dst):
    if not os.path.exists(dst):
        os.makedirs(dst)

    image_files = [f for f in os.listdir(src)
                   if os.path.isfile(os.path.join(src, f)) and os.path.splitext(f)[1].lower() in image_extensions]

    print(f"Total images found: {len(image_files)}")

    for img in image_files:
        src_path = os.path.join(src, img)
        dst_path = os.path.join(dst, img)
        shutil.move(src_path, dst_path)

    print(f"Moved {len(image_files)} images from '{src}' to '{dst}'.")

# Run the function
count_and_move_images(source_dir, destination_dir)
```

### After train-test-split i have move the images on train and val folder

```python
import os
import shutil
from tqdm import tqdm

def move_images(df, split_dir, source_dir, filename_col='filename', label_col='words'):
    """
    Move images to train/val directory with subfolders by class.

    Args:
        df: DataFrame containing image info
        split_dir: 'train' or 'val'
        source_dir: Base directory where original images are located
        filename_col: Column name in df that contains filenames
        label_col: Column name in df that contains class labels
    """

    os.makedirs(split_dir, exist_ok=True)

    for _, row in tqdm(df.iterrows(), total=len(df)):
        filename = row[filename_col]
        class_name = row[label_col]
        
        src_path = os.path.join(source_dir, filename)
        # dst_dir = os.path.join(split_dir, class_name)
        # os.makedirs(dst_dir, exist_ok=True)
        dst_path = os.path.join(split_dir, filename)
        
        try:
            shutil.copy2(src_path, dst_path)
        except FileNotFoundError:
            print(f"File not found: {src_path}")

```

```python
source_dir = '/home/arijit/LPR_project/synthetic_data_generated/generated_two_line_plates_2'

move_images(train_df, 'en_train_filtered', source_dir, filename_col='filename', label_col='words')
move_images(val_df, 'en_val', source_dir, filename_col='filename', label_col='words')
```

### Simple Label File (Format: filename words)

```python
def create_label_txt(df, output_file="label.txt"):
    """
    Create a label.txt file mapping filenames to labels.
    Format: `filename words` (space-separated).
    """
    with open(output_file, 'w') as f:
        for _, row in df.iterrows():
            f.write(f"{row['filename']} {row['words']}\n")

# For training data
create_label_txt(train_df, "/home/arijit/LPR_project/synthetic_data_generated/en_train_filtered/label.txt")

# For validation data
create_label_txt(val_df, "/home/arijit/LPR_project/synthetic_data_generated/en_val/label.txt")
```

### Python Script to Flatten a Directory Structure

```python
import os
import shutil

# Source directory containing multiple folders/home/ariji
source_dir = "/home/arijit/LPR_project/synthetic_data_generated/cropped_archive/cropped_statewise_data"

# Target directory where all files will be copied
target_dir = "/home/arijit/LPR_project/synthetic_data_generated/cropped_archive/cropped_statewise_data"
os.makedirs(target_dir, exist_ok=True)

# Traverse all folders and subfolders
for root, dirs, files in os.walk(source_dir):
    for file in files:
        if file.endswith((".jpg", ".png", ".jpeg", ".txt")):  # Add/remove extensions as needed
            src_path = os.path.join(root, file)
            dst_path = os.path.join(target_dir, os.path.basename(file))

            # If file with same name exists, rename to avoid conflict
            base, ext = os.path.splitext(file)
            counter = 1
            while os.path.exists(dst_path):
                dst_path = os.path.join(target_dir, f"{base}_{counter}{ext}")
                counter += 1

            shutil.move(src_path, dst_path)

print("All files have been extracted and copied.")
```
